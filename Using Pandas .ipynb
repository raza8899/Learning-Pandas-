```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
plt.style.use('fivethirtyeight')
import warnings
warnings.filterwarnings("ignore")
```

```python
my_data= pd.read_csv("../Python_New_Work/Pokemon.csv")
```

```python
my_data.head()
```

```python
# Making all the column name into Upper case and replacing white spaces
my_data.columns=my_data.columns.str.upper().str.replace('_','')
```

```python
my_data.head()
```

```python
my_data.columns=my_data.columns.str.replace(' ','_')
my_data.columns=my_data.columns.str.replace('.','')
```

```python
my_data.head()
```

```python
my_data=my_data.set_index('NAME')
```

```python
my_data.index= my_data.index.str.replace(".*(?=Mega)","")
```

```python
my_data.head(10)
```

```python
my_data=my_data.drop(['#'],axis=1)
```

```python
my_data.head(3)
```

```python
my_data.TYPE_2.fillna(my_data['TYPE_1'],inplace=True)
```

```python
plt.hist(x=my_data["ATTACK"],color='green')
plt.axvline(my_data.ATTACK.mean(),color='red',linestyle='dashed')
plt.xlabel('Attack')
plt.ylabel('Count')
```

```python
grass_pokemon=my_data[(my_data['TYPE_1']=='Grass') | (my_data['TYPE_2']=='Grass')]
```

```python
psychic_pokemon=my_data[(my_data['TYPE_1']=='Psychic') | (my_data['TYPE_2']=='Psychic')]
```

```python
plt.subplots(figsize=(14,7))
plt.scatter(x=grass_pokemon.ATTACK.head(40),y=grass_pokemon.DEFENSE.head(40),color='R',marker='*',
            label='Grass_Pokemon',s=50)
plt.scatter(x=psychic_pokemon.ATTACK.head(40),y=psychic_pokemon.DEFENSE.head(40),color='G',marker='^',
            label='Psychic_Pokemon',s=50)
plt.legend()
plt.xlabel('ATTACK')
plt.ylabel('DEFENSE')
```

##### Grass Pokemon have  better attack than the Psychic Pokemon 

```python
pie_data=my_data.groupby(['TYPE_1'],as_index=False).count().sort_values(by='ATTACK',ascending=False)
pie_data
```

```python
plt.subplots(figsize=(10,10))
plt.pie(x=pie_data['ATTACK'].head(10),labels=pie_data['TYPE_1'].head(10),autopct='%1.1f%%')

plt.title('Different types of pokemons')
plt.show()
```
